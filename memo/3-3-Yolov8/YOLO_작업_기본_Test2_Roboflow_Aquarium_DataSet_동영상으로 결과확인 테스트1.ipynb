```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow  # Colab ì „ìš© imshow

# YOLOv8 ëª¨ë¸ ë¡œë“œ
# model = YOLO("yolov8n.pt")  # Nano ëª¨ë¸ ì‚¬ìš©
# ì‚¬ìš©ì ì •ì˜ YOLOv8 ëª¨ë¸ ë¡œë“œ
model = YOLO("/content/drive/MyDrive/busanit501-1234/YOLOv8_Aqua/modelresult/weights/best.pt")

# ë¹„ë””ì˜¤ íŒŒì¼ ê²½ë¡œ
video_path = "/content/drive/MyDrive/busanit501-1234/YOLOv8_Aqua/á„‡á…¡á„ƒá…¡_á„†á…®á†¯á„€á…©á„€á…µ1.mp4"

# ì¶œë ¥ ë¹„ë””ì˜¤ ì„¤ì •
output_path = "/content/drive/MyDrive/busanit501-1234/YOLOv8_Aqua/output_á„‡á…¡á„ƒá…¡_á„†á…®á†¯á„€á…©á„€á…µ2.mp4"

# ë¹„ë””ì˜¤ ìº¡ì²˜ ì´ˆê¸°í™”
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Cannot open video file.")
    exit()

# ë¹„ë””ì˜¤ ì†ì„± ì½ê¸°
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*"mp4v")

# ì¶œë ¥ ë¹„ë””ì˜¤ ì´ˆê¸°í™”
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# í”„ë ˆì„ ë‹¨ìœ„ë¡œ ê°ì²´ íƒì§€ ìˆ˜í–‰
frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:  # ë¹„ë””ì˜¤ì˜ ëì— ë„ë‹¬í•œ ê²½ìš°
        break

    # YOLOv8ì„ ì‚¬ìš©í•˜ì—¬ ê°ì²´ íƒì§€
    results = model.predict(source=frame, save=False, conf=0.5)

    # íƒì§€ ê²°ê³¼ ì‹œê°í™”
    annotated_frame = results[0].plot()

    # íƒì§€ ê²°ê³¼ë¥¼ ì¶œë ¥ ë¹„ë””ì˜¤ì— ì €ì¥
    out.write(annotated_frame)

    # Colabì—ì„œ ì‹¤ì‹œê°„ í‘œì‹œ (ë§¤ 30 í”„ë ˆì„ë§ˆë‹¤ í‘œì‹œ)
    if frame_count % 30 == 0:  # ë„ˆë¬´ ìì£¼ í‘œì‹œí•˜ë©´ ì„±ëŠ¥ ì €í•˜
        cv2_imshow(annotated_frame)
    frame_count += 1

# ë¦¬ì†ŒìŠ¤ í•´ì œ
cap.release()
out.release()
print(f"Detection completed. Output video saved at {output_path}")
```
